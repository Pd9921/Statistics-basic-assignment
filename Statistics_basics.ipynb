12  Implement and compare simple random sampling and systematic sampling ?


```python
import numpy as np
population = np.arange(1, 51)
sample_size = 10
np.random.seed(0)
simple_random_sample = np.random.choice(population, size=sample_size, replace=False)
print("Simple Random Sample:", simple_random_sample)
interval = len(population) // sample_size
np.random.seed(0)
start = np.random.randint(0, interval)
systematic_sample = population[start::interval][:sample_size]
print("Systematic Sample:", systematic_sample)
```

 20 Write a Python program to visualize covariance using a heatmap


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
num_rows = 50
data = pd.DataFrame({
    'Math': np.random.randint(35, 40, num_rows),
    'Science': np.random.randint(30, 55, num_rows),
    'English': np.random.randint(85, 120, num_rows)
})
cov_matrix = data.cov()
plt.figure(figsize=(4, 5))
sns.heatmap(cov_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title("Covariance Heatmap")
plt.show()
```
